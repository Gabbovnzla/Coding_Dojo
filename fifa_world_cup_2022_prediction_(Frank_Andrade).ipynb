#Limpieza de fixture

```python
df_fixture["home"] = df_fixture["home"].str.strip()
df_fixture["away"] = df_fixture["away"].str.strip()
```

#Limpiando Missin Dat

Para verificar los Null o NaN, debemos usar isnull()

```python
df_data_faltante[df_data_faltante.isnull()]
```

```python
df_data_faltante.dropna(inplace=True)
```

```python
df_data_faltante[df_data_faltante["home"].isnull()]
```

#Ahora, concatenamos ambas datas

```python
df_data_historica= pd.concat([df_data_historica, df_data_faltante], ignore_index=True)
df_data_historica
```

#Ahora, se eliminan dupliacados

```python
df_data_historica= pd.concat([df_data_historica, df_data_faltante], ignore_index=True)
df_data_historica.drop_duplicates(inplace=True)
df_data_historica.sort_values("year", inplace=True)
df_data_historica
```

Dentro del "score" deberian estar algunos parametros con letras, las cuales debemos limpiar de esta manera.

```python
df_data_historica["score"].str.contains("[^\d-]")
```

```python
df_data_historica[df_data_historica["score"].str.contains("[^\d–]")]
```

```python
df_data_historica["score"] = df_data_historica["score"].str.replace("[^\d-]", " ", regex = True)
df_data_historica["score"]
```

```python
df_data_historica[df_data_historica["score"].str.contains("[^\d–]")]
```

```python
df_data_historica["home"] = df_data_historica["home"].str.strip()
df_data_historica["away"] = df_data_historica["away"].str.strip()
```

#Ahora, vamos a separar el score en dos columnas con el metodo split()

```python
df_data_historica["score"].str.split(" – ")
```

```python
df_data_historica["score"].str.split(" – ", expand = True) #para ver cuantas columnas de score tengo
```

Renombramos las columnas para que aparezcan

```python
df_data_historica["HomeGoals", "AwayGoals"] = df_data_historica["score"].str.split(" – ", expand = True)
```

```python
df_data_historica.drop('score', axis=1, inplace=True)
```

```python
df_data_historica
```

#Para poder renombrar las primeras letras en mayusculas, se hace lo siguiente:

```python
df_data_historica.rename(columns={'home': 'HomeTeam', 'away': 'AwayTeam',
                                   'year':'Year'}, inplace=True)
```
